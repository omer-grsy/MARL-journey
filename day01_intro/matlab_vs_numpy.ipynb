# MATLAB → NumPy Cheat Sheet

Bu notebook, MATLAB kullananlar için NumPy/Python karşılıklarını hızlıca öğrenmek amacıyla hazırlanmıştır.

Amaç:
- Temel syntax farkları
- Indexing
- Linear algebra
- Broadcasting mantığı

## Array oluşturma

```python
# MATLAB
# A = [1 2; 3 4]

# NumPy
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

print(A)
```

## Indexing (ÇOK ÖNEMLİ)

MATLAB: 1-based  
NumPy: 0-based

```python
A = np.array([[1, 2],
              [3, 4]])

# MATLAB: A(1,1)
print(A[0, 0])  # NumPy
```

## Slicing

```python
# MATLAB: A(:,1)
print(A[:, 0])  # tüm satırlar, 1. kolon

# MATLAB: A(1,:)
print(A[0, :])  # 1. satır
```

## Element-wise vs Matrix operations

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Element-wise (MATLAB: .*)
print(A * B)

# Matrix multiply (MATLAB: *)
print(A @ B)
```

## Transpose

```python
print(A.T)
```

## Shape & Size

```python
print(A.shape)   # (2,2)
print(A.size)    # 4
```

## Loop yerine vectorization kullan

```python
# KÖTÜ
result = []
for x in range(5):
    result.append(x**2)

# İYİ
arr = np.arange(5)
print(arr ** 2)
```

## Broadcasting

NumPy otomatik genişletir (MATLAB'da daha sınırlı)

```python
A = np.array([[1, 2],
              [3, 4]])

b = np.array([10, 20])

print(A + b)
```

## Random

```python
# MATLAB: rand(2,2)
print(np.random.rand(2,2))
```

## Linear Algebra

```python
A = np.array([[1, 2],
              [3, 4]])

# determinant
print(np.linalg.det(A))

# inverse
print(np.linalg.inv(A))

# eigenvalues
print(np.linalg.eig(A))
```

## Özet

- MATLAB → 1-based index
- NumPy → 0-based index
- MATLAB → matrix default
- NumPy → array default
- MATLAB → * = matrix multiply
- NumPy → * = element-wise, @ = matrix multiply
- NumPy → broadcasting güçlü
